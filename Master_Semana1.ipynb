# Tipos de Variáveis

Variável é uma característica de interesse que é medida em cada elemento de uma população ou amostra. As variáveis se dividem em diferentes tipos:

1. **Quantitativas**: representam uma quantidade, ou seja, um valor numérico. Podem ser divididas em:

  - **Discretas**: Só podem assumir um número finito ou infinito contável de valores. Exemplo: número de filhos, número de carros em uma cidade, número de pessoas que vivem em um prédio.

  - **Contínuas**: Assumem valores numéricos dentro de um intervalo contínuo. Exemplo: altura de uma pessoa, tempo, distância.

2. **Qualitativas (ou categóricas)**: representam uma qualidade e não uma quantidade. Podem ser divididas em:

  - **Nominais**: Não existe uma ordem no valores que a variável pode assumir. Exemplo: sexo, tipo sanguíneo, time de preferência.

  - **Ordinais**: Existe uma ordem natural para os valores que a variável pode assumir. Exemplo: escolaridade, mês de observação.

Levando em consideração as definições anteriores, vale ressaltar que nem sempre as variáveis são representadas de uma única forma. Por exemplo, a variável altura é quantitativa contínua se considerada da forma usual mas pode ser qualitativa ordinal se criarmos faixas como, por exemplo, altura menores que 1,50 m (baixo); alturas entre 1,50 m e 1,80 m (mediano) e acima de 1,80 m (alto). Dessa forma teríamos 3 categorias de alturas.

Além disso, nem sempre uma variável que apresenta número é quantitativa. Por exemplo, poderíamos representar sexo com os valores 0 e 1 ao invés de M e F. Essa variável embora apresente número é ainda qualitativa nominal.

Digamos que nós estamos interessados em analisar a variável `Densidade demográfica` que é quantitativa contínua mas nosso interesse seja olhar para localidade que possuam tal densidade abaixo de 100 ou acima de 100 para efeitos de identificar possíveis municípios que possuam grande mercado consumidor para determinado produto.

Como poderíamos transformar essa variável?

Uma forma é criar uma função que nos auxilie nessa tarefa e depois usá-la para criar uma nova coluna do `DataFrame`.

```python
# função que cria uma coluna categórica a partir de valores numéricos
def dens_dem(df, coluna, limite):
  cat = []
  for valor in df[coluna]:
    if valor <= limite:
      cat.append(0)
    else:
      cat.append(1)

  return cat
```

```python
# criando a coluna densidade demográfica categórica
df['Densidade demográfica categórica'] = dens_dem(df, 'Densidade demográfica', 100)

# visualizando o DataFrame
df.head()
```

Por padrão, a nova coluna criada fica no final do `DataFrame`. Vamos criar uma função que nos permita reordenar as colunas.

```python
def reordena_colunas(df, col_name, index):
    # criando uma lista com o nome de todas as colunas do DataFrame
    cols = df.columns.tolist()

    # removendo a coluna que vamos reordenar
    cols.remove(col_name)

    # inserindo a coluna removida na posição desejada
    cols.insert(index, col_name)

    # retorna o DataFrame ordenado
    return df[cols]
```

```python
# aplicando a função de reordenar colunas ao nosso DataFrame
df = reordena_colunas(df, 'Densidade demográfica categórica', 4)
df.head()
```

# Correlação

Correlação é uma maneira de mensurar o quanto duas variáveis estão relacionadas/associadas uma com a outra. Há várias maneiras de medir essa associação, isto depende se as variáveis são numéricas e/ou categóricas. O coeficiente de correlação varia de -1 a +1.

A correlação vai retornar três possíveis casos:

* correlação positiva: quando duas variáveis crescem ou decrescem juntas, ou seja, quando uma cresce a outra também cresce e quando uma decresce a outra também decresce;

* correlação negativa: quando uma variável cresce a outra decresce, ou vice-versa;

* não há correlação: quando o crescimento ou decrescimento de uma variável não tem efeito sobre outra variável.

Calcular correlação usando `pandas` é bem simples, porque essa função já está implementada na biblioteca.

## Correlação de Pearson

Mesmo entre variáveis numéricas existe mais de uma forma de medir a correlação e a correlação de Pearson mede a existência de uma associação linear entre duas variáveis numéricas.

Se a correlação for exatamente 1 ou -1 significa que existe uma relação linear exata entre as variáveis (y = 2x, por exemplo). Coeficiente igual ou próximo de zero indica qua não há correlação linear.

**Observação**: O coeficiente de correlação de Pearson mede apenas correlação **linear**, então tome cuidado ao afirma que duas variáveis não estão correlacionadas caso o coeficiente de Pearson seja próximo de zero, pois pode haver uma correlação não linear nos dados!

Para ter uma ideia se há correlação linear entre duas variáveis pode-se plotar um gráfico de dispersão. Outro pressuposto é que os dados devem estar normalmente distribuídos.

## Correlação de Spearman

O coeficiente de correlação de Spearman não precisa que os dados sejam normalmente distribuídos e consegue medir correlações não lineares. Portanto, se seus dados não forem numéricos este é um coeficiente de correlação interessante de ser utilizado.

## Correlação com `pandas`

Vamos usar função do `pandas` para calcular a correlação entre as variáveis `Taxa de Urbanização` e `Taxa de Saneamento Básico`.

Vamos importar a biblioteca `seaborn` para fazer tal visualização e utilizar a função `scatterplot()` para observar se há indícios de relação linear entre as variáveis.

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df, x='Taxa de Urbanização', y='Taxa de Saneamento Básico');
```

Podemos notar que há grande dispersão dos dados, não aparentando uma correlação linear muito forte embora possamos notar que para valores maiores de `Taxa de Urbanização` há uma amplitude maior de valores (consequentemente maiores valores) assumidos pela variável `Taxa de Saneamento Básico`, portanto não devemos esperar encontrar uma correlação muito alta.

```python
df[['Taxa de Urbanização', 'Taxa de Saneamento Básico']].corr()
```

Podemos observar que na tabela resultante a correlação de uma variável consigo mesma é igual a 1, isso sempre acontecerá. Nosso interesse está em analisar a correlação de uma variável com a outra e neste caso o valor é de aproximadamente 0.55.

Por padrão o `pandas` usa a correlação de Pearson, por isso devemos especificar um parâmetro caso nosso interesse seja em outro tipo de correlação, conforme código a seguir.

```python
df[['Taxa de Urbanização', 'Taxa de Saneamento Básico']].corr(method='pearson')
```

```python
df[['Taxa de Urbanização', 'Taxa de Saneamento Básico']].corr(method='spearman')
```

```python
df[['Taxa de Urbanização', 'Taxa de Saneamento Básico']].corr(method='kendall')
```

Embora seja bem simples calcular essa formas de correlação, deve-se ter em mente qual é a forma mais adequada para o problema em questão.

E se quisessemos visualizar a correlação entre várias variáveis? É só usar a mesma função do `pandas`.

```python
df.corr(method='pearson')
```

Neste caso o pandas ignora a variáveis não numéricas para conseguir fazer as contas. Mas perceba que visualizar uma tabela numérica como essa não é nada agradável. Como poderíamos entender melhor a correlação entre várias variáveis? Uma maneira interessante é o mapa de calor (heatmap).

Vamos passar para a função `heatmap()` do `seaborn` uma matriz de correlações do `pandas`, o valor mínimo e o valor máximo que estamos interessados. Podemos ainda usar o parâmetro `annot` para visualizar os valores da correlação no mapa de calor, mas no nosso caso (devido ao nome das variáveis) é mais interessante deixar esse parâmetro igual a `False`.

```python
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=False);
```

Podemos ver que há alguns quadrado muito próximos do branco, indicando forte correlação positiva, e alguns muito próximos do preto, indicando forte correlação negativa.


Para mais informações sobre a função `heatmap()` da biblioteca `seaborn` consulte [heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
